Import necessary parts for making plots and data analysis

```python
import pandas as pd
import matplotlib.pyplot as plot
import seaborn as sns
import datetime as dt
```

Read in the data from our group

```python
data_df = pd.read_csv("https://smarkwardt.greenriverdev.com/DataProject/SurveyData.csv")
# make it print out to be sure it imported correctly
data_df
```

Check the correlation table - we are missing our days to overshoot column since it is not int type

```python
data_df.corr()
```

Turn the date column into a date object as step one to making it an int type

```python
data_df['Personal Earth Overshoot Day'] = pd.to_datetime(data_df['Personal Earth Overshoot Day'], format = '%m/%d')
data_df
```

Calculate the days to overshoot to use in our table

```python
delta = dt.datetime(1900, 1, 1)

data_df['Days to Overshoot'] = data_df['Personal Earth Overshoot Day'] - delta
data_df
```

Turn the days into an int type

```python
data_df['Days to Overshoot'] = pd.to_timedelta(data_df['Days to Overshoot']).dt.days
data_df
```

Now we can make our final correlation table. We'll use a heatmap so it is easier to read and we will only use the collumns we need

```python
# select the collumns
needed_df = data_df[['Days to Overshoot', 'Food -gha', 'Shelter - gha', 'Mobility - gha', 'Goods - gha', 'Service - gha']]

# now we can make our heatmap
plot.figure(figsize=(8, 8))
sns.heatmap(needed_df.corr(),annot=True)
```
